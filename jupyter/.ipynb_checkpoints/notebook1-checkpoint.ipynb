```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

# STEP 1 Load dataset

```python
covid = pd.read_csv('../input/covid-19.csv')
```

# STEP 2 Exploring data

```python
covid.tail(10)
```

```python
print(covid.columns)
```

# STEP 3 Feature reduction

```python
covid = covid.drop(['No.', 'Notified date', 'Unit', 'province_of_onset',
                    'district_of_onset', 'Unnamed: 11', 'Unnamed: 12',
                    'Unnamed: 13'], axis = 1)
```

```python
covid.tail(10)
```

```python
(covid.province_of_isolation).unique()
```

```python
(covid.nationality).unique()
```

```python
(covid.risk).unique()
```

# STEP 4 Cleansing data

```python
(covid.isnull()).sum()
```

```python
covid = covid.dropna(axis = 0, subset = ['sex', 'age',
                                         'nationality', 'province_of_isolation',
                                         'risk'])
(covid.isnull()).sum()
```

# STEP 5 Prepare data (re-format)

```python
covid.dtypes
```

```python
covid['announce_date'] = pd.to_datetime(covid['announce_date'], format = '%d/%m/%Y')
```

```python
covid.dtypes
```

```python
covid['announce_y'] = covid['announce_date'].dt.year
covid['announce_m'] = covid['announce_date'].dt.month
covid['announce_d'] = covid['announce_date'].dt.day
```

```python
covid = covid.drop(['announce_date'], axis = 1)

covid.tail(10)
```

```python
covid.loc[(covid["announce_y"] == 2021)]
```

```python
covid.loc[(covid.index >= 120) & (covid.index <= 130)]
```

```python
covid['person_count'] = 1

covid.tail(10)
```

```python
covid_2021_04 = covid.loc[(covid["announce_y"] == 2021) & (covid["announce_m"] >= 4)]
covid_2021_04_05 = covid_2021_04.loc[(covid["announce_m"] == 4) | ( (covid["announce_m"] == 5) & (covid["announce_d"] <= 5) )]
covid_2021_04_05
```

```python
covid_2021_04_05 = pd.pivot_table(covid_2021_04_05,
                                  index = ["announce_y","announce_m","announce_d"],
                                  values = ["person_count"],
                                  aggfunc = {"person_count":np.sum})

covid_2021_04_05 = covid_2021_04_05.reset_index()
covid_2021_04_05
```

```python
covid_2021_04_05.index = covid_2021_04_05.index + 1
covid_2021_04_05
```

```python
plt.figure(figsize = (15, 7))
sns.lineplot(covid_2021_04_05.index, 'person_count', data = covid_2021_04_05, marker = 'o', markersize = 10)
plt.show
```

# STEP 6 Training data

```python
x = ((covid_2021_04_05.index).values).reshape(-1,1)    # features

y = covid_2021_04_05.person_count                      # labels
```

```python
poly_features = PolynomialFeatures(degree = 2)
x_poly = poly_features.fit_transform(x)            # fit data
model = LinearRegression()
model.fit(x_poly, y)                               # train data
```

```python
y_poly_pred = model.predict(x_poly)
```

# STEP 7 Plot chart

```python
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = 15,7
plt.scatter(x, y, color = 'b', s = 12)
plt.plot(x, y_poly_pred, linewidth = '2', color = 'r')       # model line
plt.ylabel('New cases')
plt.xlabel('date')
plt.title('degee15')
plt.show()
```

# STEP 8 Evaluation

```python
print('r2 = {:0.5f}'.format(r2_score(y, y_poly_pred)))
```

# STEP 9 Prediction

```python
x_input = [[40], [41], [42]]
pred = model.predict(poly_features.fit_transform(x_input))
```
