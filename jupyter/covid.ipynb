# ➡️ Remove Features

---

```python
covid = covid.drop(['No.', 'Notified date', 'Unit', 'province_of_onset', 'district_of_onset'], axis = 1)
```

```python
covid.tail(10)
```

---

# ➡️ Cleansing data

---

```python
(covid.isnull()).sum()         # see number of NULL each column
```

```python
covid = covid.dropna(axis = 0, subset = ['sex', 'age',
                                         'nationality', 'province_of_isolation',
                                         'risk'])
(covid.isnull()).sum()
```

---

# ➡️ Preparing (re-format)

---

```python
covid.dtypes          # see type
```

```python
covid['announce_date'] = pd.to_datetime(covid['announce_date'], format = '%d/%m/%Y')    # convert variable type
```

```python
covid.dtypes
```

```python
# separate year, month and day

covid['announce_y'] = covid['announce_date'].dt.year
covid['announce_m'] = covid['announce_date'].dt.month
covid['announce_d'] = covid['announce_date'].dt.day
```

```python
covid = covid.drop(['announce_date'], axis = 1)

covid.tail(10)
```

```python
covid.loc[(covid["announce_y"] == 2021)]
```

```python
covid.loc[(covid.index >= 120) & (covid.index <= 130)]      # see a outlier
```

```python
covid['person_count'] = 1

covid.tail(10)
```

```python
covid_2021_04 = covid.loc[(covid["announce_y"] == 2021) & (covid["announce_m"] >= 4) & (covid["announce_m"] <= 6)]
covid_2021_04
```

```python
covid_2021_04 = pd.pivot_table(covid_2021_04,
                               index = ["announce_y","announce_m","announce_d"],
                               values = ["person_count"],
                               aggfunc = {"person_count":np.sum})

covid_2021_04 = covid_2021_04.reset_index()
covid_2021_04
```

```python
covid_2021_04.index = covid_2021_04.index + 1       # change index
covid_2021_04
```

---

# ➡️ Plot chart (pre training)

---

```python
plt.figure(figsize = (15, 7))
sns.lineplot(covid_2021_04.index, 'person_count', data = covid_2021_04, marker = 'o', markersize = 10)
plt.show
```

---

# ➡️ Training data

---

```python
x = ((covid_2021_04.index).values).reshape(-1,1)    # features

y = covid_2021_04.person_count                      # labels
```

```python
poly_features = PolynomialFeatures(degree = 5)
x_poly = poly_features.fit_transform(x)            # fit data
model = LinearRegression()
model.fit(x_poly, y)                               # train data
```

```python
y_poly_pred = model.predict(x_poly)
```

---

# ➡️ Plot chart

---

```python
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = 15,7
plt.scatter(x, y, color = 'b', s = 12)
plt.plot(x, y_poly_pred, linewidth = '2', color = 'r')       # model line
plt.ylabel('New cases')
plt.xlabel('Date')
plt.title('degee15')
plt.show()
```

---

# ➡️ Evaluation

---

```python
print('r2 = {:0.5f}'.format(r2_score(y, y_poly_pred)))
```

---

---
